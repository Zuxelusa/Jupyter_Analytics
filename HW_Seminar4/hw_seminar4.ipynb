#### 1.1. Данные из прошлого дз



```python
import pandas as pd
df = pd.read_csv('kc_house_data.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

#### 1.2 Изучите стоимости недвижимости


```python
plt.hist(df['price'], bins=50);
plt.title('Распределение объектов по стоимости недвижимости')
plt.xlabel('Стоимость объекта недвижимости, млн.руб.')
plt.ylabel('Количество объектов недвижимости')
```

```python
sns.boxplot(x=df['price'], showfliers = False)
plt.title('Распределение объектов по стоимости недвижимости')
plt.xlabel('Стоимость объекта недвижимости, млн.руб.')
```

```python
sns.boxplot(x=df['price'], showfliers = True)
plt.title('Распределение объектов по стоимости недвижимости')
plt.xlabel('Стоимость объекта недвижимости, млн.руб.')
```

**Вывод: из полученной визуализации мы видим что стоимость объектов недвижимости преимущественно находится в диапазоне цен от 300 до 650 тыс. при медиане около 450 тыс. В совокупности имеются так называемые "выбросы" - более дорогое жилье, отмечены черными точками на последнем графике.**

#### 1.3 Изучите распределение квадратуры жилой


```python
plt.hist(df['sqft_living'], bins='auto');
plt.title('Распределение объектов по площади жилой')
plt.xlabel('Жилая площадь, кв.фут')
plt.ylabel('Количество объектов недвижимости')
```

```python
sns.boxplot(x=df['sqft_living'], showfliers = True)
plt.title('Распределение объектов по стоимости недвижимости')
plt.title('Распределение объектов по площади жилой')
plt.xlabel('Жилая площадь, кв.фут')
```

```python
sns.boxplot(x=df['sqft_living'], showfliers = False)
plt.title('Распределение объектов по стоимости недвижимости')
plt.title('Распределение объектов по площади жилой')
plt.xlabel('Жилая площадь, кв.фут')
```

```python
sns.jointplot(x=df['price'], y=df['sqft_living'],kind='reg');
```

**Вывод: основное количество объектов - в диапазоне жилой площади от 1300 до 2600. Распределение жилой площади объектов недвижимости в совокупности находится в полной корелляции с его стоимостью, чем больше площадь, тем выше цена. при этом мы видим, что распределение при увеличении стоимости расходятся в стороны, что дает нам понимание о наличии дополнительно других факторов влияющих на конечную стоимость. График "ящик с усами" для стоимости идентичен графику для жилой площади.**

#### 1.4 Изучите распределение года постройки


```python
sns.histplot(df['yr_built'], bins=120);
plt.title('Объемы строительства объектов по годам')
plt.xlabel("Год постройки")
plt.ylabel("Кол-во построенных объектов");
```

```python
sns.boxplot(x=df['yr_built'], showfliers = True)
plt.title('Объемы строительства объектов по годам')
plt.xlabel('Год постройки')
```

**Вывод: строительство объектов велось за весь период стабильными темпами, мы не наблюдаем выбросов в совокупности. Как мы видим, были годы, когда стройка не велась совсем, 4 периода с равным удалением друг от друга. Основной период строительства объектов с 1950 по 1995 гг.**

```python
round(df.describe(), 1)
```

```python
df_corr = round(df.corr(), 1)
df_corr
```

```python
plt.figure(figsize=(10, 10))
sns.heatmap(df_corr, annot=True, cmap='coolwarm')
plt.title('Таблица корреляции признаков объектов')
```

**Вывод: наиболее коррелируемые с ценой признаки обхектов, это Жилая площадь, престижность объекта, надземная площадь, среднее влияние оказывает количество санузлов и вид из окна. Проанализируем эти зависимости**

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg');
```

**Вывод: прослеживается зависмость цены от жилой площади, при это при увеличении одного из показателей, распределение расширяется, что оворит нам о наличие комплекса факторов влияющих на стоимость помимо площади.**

```python
sns.boxplot(x=df['view'], y=df['price'])
plt.title('Распределения стоимости объектов в зависмости от наличия вида из окна')
```

**Вывод: Из графика видно, что стоимость объектов зависит от качества вида из окна**

```python
sns.barplot(x=df['bathrooms'], y=df['price']);
plt.title("Зависимость цены объекта от количества ванных комнат")
plt.xticks(rotation=50);
```

**Вывод: Из графика видно, что стоимость объектов зависит от количества санузлов**
